```python
import matplotlib.pyplot as plt
```

```python
moco_log = './logs/MoCo_ResNet50_200.txt'
```

```python
accs = []
lrs = []
train_losses = []
val_losses = []

with open(moco_log, 'r') as l:
    lines = l.readlines()
    for i, line in enumerate(lines):
        line = line[:-1]
        if i > 26:
            infos = line.split(',')
            accs.append(float(infos[1]))
            lrs.append(float(infos[2]))
            train_losses.append(float(infos[3]))
            val_losses.append(float(infos[5]))
```

```python
label_fontsize = 25
```

```python
plt.figure(figsize=(20, 20))
plt.axis('off')

plt.subplot(2, 1, 1); plt.grid()
acc_line, = plt.plot(accs)
plt.xlabel('Epoch', fontsize=label_fontsize)
plt.ylabel('Acc', fontsize=label_fontsize)

plt.subplot(2, 1, 2); plt.grid()
lr_line, = plt.plot(lrs)
plt.xlabel('Epoch', fontsize=label_fontsize)
plt.ylabel('LR', fontsize=label_fontsize)
```

```python
plt.figure(figsize=(20, 10))
plt.grid()

train_lossline, = plt.plot(train_losses, label='train')
val_lossline, = plt.plot(val_losses, color='red', label='val')
plt.legend(handles=[train_lossline, val_lossline], fontsize=label_fontsize)
plt.xlabel('Epoch', fontsize=label_fontsize)
plt.ylabel('Loss', fontsize=label_fontsize)
```
